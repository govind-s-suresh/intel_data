```python
from __future__ import print_function
import os
data_path = ['.']
```

```python
import pandas as pd
import numpy as np
# Import the data using the file path
filepath = os.sep.join(data_path + ['Mobile-Phones.csv'])
data = pd.read_csv(filepath)
data
```

```python
data.head(1).T
```

```python
data.drop(['main_camera', 'resolution', 'sim_card','region','selfie_camera(MP)','location','storage(GB)','sd_card','brand', 'model','display','color','os'], axis=1, inplace=True)
```

```python
x_cols = [x for x in data.columns if x != 'price(¢)']
data[x_cols] = data[x_cols].fillna(0)


#print(x_cols)
# Split the data into two dataframes
X_data = data[x_cols]
y_data = data['price(¢)']
print(X_data)
print(y_data)

```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_data)
```

```python
def accuracy(real, predict):
    return sum(y_data == y_pred) / float(real.shape[0])
```

```python
print(accuracy(y_data, y_pred))
```
